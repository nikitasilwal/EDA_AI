```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/drive/MyDrive/eda_nikita_lab/Uncleaned_employees_final_dataset (1).csv")
df.head(5)
```

```python
# Categorical data
# countplot
sns.countplot(data=df, x='gender')
df['employee_id'].value_counts()
```

```python
 # pie chart

df['gender'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df['age'])
```

```python
# Distplot
# probability density function --> The line is kde
# skweness
sns.distplot(df['age'])
```

```python
# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['age'])
```

```python
df.describe()
```

```python
# Scatter plot(Numerical - Numerical) for the relation between two column
sns.scatterplot(x='no_of_trainings', y='age',hue='gender', style='region',size='department' ,data=df)
plt.show()
```

```python
  # Bar plot (Numerical - Categorical)
  # average age in pclass
sns.barplot(x='gender',y='age',hue='department',data=df)
```

```python
#Box plot (Numerical - Categorical)

sns.boxplot(x='age',y='gender',hue='department',data=df)
```

```python
# HeatMap (Categorical - Categorical)
sns.heatmap(pd.crosstab(df['age'],df['gender']))
```

```python
# pairplot
sns.pairplot(df,hue='gender')
```
